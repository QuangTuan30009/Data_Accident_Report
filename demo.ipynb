Accident Severity by Day of the Week

```python
import pandas as pd
import matplotlib.pyplot as plt

# Định nghĩa thứ tự đúng của các ngày trong tuần
day_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
df['Day_of_Week'] = pd.Categorical(df['Day_of_Week'], categories=day_order, ordered=True)

# Nhóm dữ liệu và vẽ biểu đồ
severity_by_day = df.groupby(['Day_of_Week', 'Accident_Severity']).size().unstack(fill_value=0)
severity_by_day_percentage = severity_by_day.div(severity_by_day.sum(axis=1), axis=0) * 100

# Vẽ biểu đồ theo thứ tự đúng
severity_by_day_percentage.plot(kind='bar', stacked=True, figsize=(10, 6))
plt.title('Accident Severity by Day of the Week')
plt.xlabel('Day of the Week')
plt.ylabel('Percentage of Accidents')
plt.legend(title='Accident Severity')
plt.show()
```

Biểu đồ hiển thị tỷ lệ phần trăm các vụ tai nạn theo mức độ nghiêm trọng (Fatal, Serious, Slight) trên từng ngày trong tuần.

1️⃣ Những điều nhận thấy từ biểu đồ:
✅ Hầu hết tai nạn đều thuộc mức "Slight" (nhẹ)

Màu xanh lá (Slight) chiếm tỷ lệ lớn nhất trên tất cả các ngày.
Điều này cho thấy đa phần tai nạn không quá nghiêm trọng.
✅ Tỷ lệ tai nạn nghiêm trọng (Serious - màu cam) cao hơn vào thứ Hai và Chủ Nhật

Có thể do:
Thứ Hai: Đầu tuần, giao thông đông đúc, áp lực đi làm.
Chủ Nhật: Có thể liên quan đến các chuyến đi chơi xa, lái xe không cẩn thận hoặc ảnh hưởng từ rượu bia.
✅ Tai nạn tử vong (Fatal - màu xanh dương) xuất hiện ít nhưng không bằng 0

Dù rất nhỏ, nhưng vẫn có mặt trên tất cả các ngày.
Nếu muốn giảm số vụ tai nạn chết người, cần phân tích kỹ các yếu tố đường sá, tốc độ, điều kiện lái xe ban đêm.


1️⃣ Nhấn mạnh vào ngày nguy hiểm nhất

Nếu dữ liệu cho thấy Chủ Nhật có nhiều tai nạn nghiêm trọng nhất, có thể đề xuất các biện pháp kiểm tra nồng độ cồn, hạn chế tốc độ trên đường cao tốc vào cuối tuần.
Nếu thứ Hai nguy hiểm, có thể đề xuất quản lý giao thông vào giờ cao điểm.
2️⃣ Hướng đến giảm tai nạn nghiêm trọng & tử vong

Vì tai nạn nhẹ là chủ yếu, nhưng tai nạn nghiêm trọng vẫn có một tỷ lệ đáng kể, nên cần kiểm soát tốc độ và xử lý hành vi vi phạm giao thông.
3️⃣ Đề xuất giải pháp dựa trên dữ liệu

Nếu mục tiêu là giảm tai nạn tử vong, có thể tập trung vào các ngày có tỷ lệ "Fatal" cao nhất.
Nếu muốn giảm tai nạn nghiêm trọng, nên kiểm soát tốt hơn vào cuối tuần và giờ cao điểm.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Giả sử df có cột 'Accident_Date' và 'Accident_Severity'
df['Month'] = pd.to_datetime(df['Accident Date']).dt.month

# Nhóm theo tháng và mức độ nghiêm trọng
severity_by_month = df.groupby(['Month', 'Accident_Severity']).size().unstack(fill_value=0)

# Vẽ biểu đồ cột stacked theo số lượng thực tế
severity_by_month.plot(kind='bar', stacked=True, figsize=(10, 6), color=['blue', 'gray', 'red'])

# Thiết lập tiêu đề và nhãn
plt.title('Number of Accidents by Severity and Month')
plt.xlabel('Month')
plt.ylabel('Number of Accidents')
plt.xticks(ticks=range(12), labels=['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
plt.legend(title='Accident Severity')

# Hiển thị biểu đồ
plt.show()
```
